```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv("titanic.csv")
df
```

```python
#ML Prediction based upon Pclass,Sex,Age,Fare
#Target Variable is Survived where 0=not survived and 1=survived
```

```python
data=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis=1)
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Nominal Data conversion
gender=pd.get_dummies(data.Sex).iloc[:,:1]
gender
```

```python
data=pd.concat([data,gender],axis=1)
data
```

```python
new_data=data.drop(['Sex'],axis=1)
new_data
```

```python
#Data imputation
new_data.isnull()
new_data.isnull().sum()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.impute import KNNImputer
```

```python
#guessing n-neighborhood to be 2 to avoid straying valuation error
impute=KNNImputer(n_neighbors=2)
data1= impute.fit_transform(new_data)
```

```python
ndata=pd.DataFrame(data1,columns=['Survived','Pclass','Age','Fare','female'])
ndata
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
nd1 = ndata.rename(columns={'female': 'Gender'})
nd1
```

```python
#Defining
x=nd1[["Pclass","Age","Fare","Gender"]]
y=nd1["Survived"]
```

```python
#Re-applying "fixed"-randomized 80:20 split model
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=10)
nd1
```
